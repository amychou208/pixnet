```python
# coding:utf-8
#from selenium import webdriver
#from selenium.webdriver.common.by import By
#from selenium.webdriver.common.keys import Keys
#from selenium.webdriver.support.ui import Select
#from selenium.common.exceptions import NoSuchElementException
#from selenium.common.exceptions import NoAlertPresentException
#import unittest, time, re
import requests as r
from bs4 import BeautifulSoup

#driver = webdriver.Chrome()
#driver.implicitly_wait(3)

page = "https://www.pixnet.net/blog/articles/category/26/hot/1"
res = r.get(page)
res.encoding='UTF-8'
soup = BeautifulSoup(res.text, 'lxml')

url_list = []

#第一名url
first_url = soup.select_one('div.featured').select_one('a')['href']
url_list.append(first_url)
#print url_list 

#第二名以後url
for i in range (0,9):
    try:
        url = soup.select_one('ol.article-list').select('li')[i].select_one('a')['href']
        url_list.append(url)
    except Exception as e:
        print e
#print url_list 

#第二頁以後url
for j in range (2,3):
    page2 = "https://www.pixnet.net/blog/articles/category/26/hot/{}"
    res2 = r.get(page2.format(j))
    res2.encoding='UTF-8'
    soup2 = BeautifulSoup(res2.text, 'lxml')
    for l in range (0,10):
        try:
            urls = soup2.select_one('ol.article-list').select('li')[l].select_one('a')['href']
            url_list.append(urls)
        except Exception as e:
            print e
print url_list

#driver.find_element_by_link_text(u"下一頁").click()
```

# use url crawling posts

```python
# coding:utf-8

import requests as r
import lxml
import json
import time
from bs4 import BeautifulSoup

# print url_list

post_list = []

for p in range (0,len(url_list)):
    post_url = url_list[p]
    res3 = r.get(post_url)
    res3.encoding='UTF-8'
    soup3 = BeautifulSoup(res3.text, 'lxml')

    #print "Crawling {}".format(post_url)
    try:
        post_dict = {}
        post_dict['url'] = post_url
        post_dict['title'] = soup3.select_one('ul.article-head').text.replace('\n','')
        post_dict['date'] = soup3.select_one('li.publish').text
        post_dict['content'] = soup3.select_one('div.article-content-inner').text.replace('\n','')
        post_dict['tag']  = soup3.select_one('div.article-keyword').text.replace('\n','')
    except Exception :
        post_dict['tag'] = 'no tag'        
        
    post_list.append(post_dict)

for post in post_list:
    for k in post.items():
        print k
        
pixnetpost = json.dumps(post_list, ensure_ascii=False)
with open('pixnetpost.json', 'w') as a:
    a.write(pixnetpost.encode('utf-8'))        
```
